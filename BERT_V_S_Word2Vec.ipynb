```python
import numpy as np
from flair.embeddings import WordEmbeddings
from flair.embeddings import TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

```python
glove_embedding=WordEmbeddings('glove')
```

```python
sentence_1=Sentence('apple released iphone 12 pro max in 2020')
```

```python
glove_embedding.embed(sentence_1)
```

```python
for token in sentence_1:
  print(token)
  print(token.embedding)
  print("\n")
```

```python
sentence_1[0]
```

```python
sentence_1[0].embedding
```

```python
sentence_1[0].embedding.shape
```

```python
sentence_2=Sentence('an apple a day keeps the doctor away')
```

```python
glove_embedding.embed(sentence_2)
```

```python
for token in sentence_2:
  print(token)
  print(token.embedding)
  print("\n")
```

```python
sentence_2[1]
```

```python
sentence_2[1].embedding
```

```python
sentence_2[1].embedding.shape
```

```python
import torch

# Assuming sentence_1[0].embedding and sentence_2[1].embedding are PyTorch tensors
embedding_1_cpu = sentence_1[0].embedding.cpu()
embedding_2_cpu = sentence_2[1].embedding.cpu()

# Convert to NumPy arrays
embedding_1_np = embedding_1_cpu.numpy()
embedding_2_np = embedding_2_cpu.numpy()

# Calculate Euclidean distance
glove_dist = distance.euclidean(embedding_1_np, embedding_2_np)
```

```python
glove_dist
```

Word2Vec embeddings lack contextual information, leading to identical embeddings for the word "apple" in different sentences, despite differing meanings. BERT embeddings, on the other hand, address this limitation by providing contextualized embeddings, capturing nuances in meaning based on surrounding context.
