```python
import pandas as pd
import numpy as np
from copy import deepcopy
from bertopic import BERTopic
```

```python
df=pd.read_csv("/content/drive/MyDrive/csv/Ola Customer Reviews.csv")
```

```python
df.head(3)
```

```python
docs=(df.loc[0:1000,"review_description"])
```

```python
docs[:5]
```

```python
model=BERTopic(language="english")
```

```python
for i, doc in enumerate(docs):
    if isinstance(doc, float):
        print(f"Index {i} contains a float value: {doc}")
```

```python
docs.dropna(inplace=True)
docs = docs.values.tolist()
```

```python
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords if not already downloaded
nltk.download('stopwords')

# Get the English stopwords from NLTK
stop_words = set(stopwords.words('english'))

# List to store documents without stopwords
docs_without_stopwords = []

# Loop through each document in your 'docs' list
for doc in docs:
    # Tokenize the document into words
    words = doc.split()
    # Remove stopwords from the document
    filtered_words = [word for word in words if word.lower() not in stop_words]
    # Join the filtered words back into a document
    filtered_doc = ' '.join(filtered_words)
    # Append the document without stopwords to the new list
    docs_without_stopwords.append(filtered_doc)

# Now 'docs_without_stopwords' contains the documents with stopwords removed
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

# Download NLTK stopwords and WordNet if not already downloaded
nltk.download('stopwords')
nltk.download('wordnet')

# Get the English stopwords from NLTK
stop_words = set(stopwords.words('english'))

# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# List to store documents without stopwords and with lemmatized words
docs_without_stopwords_lemmatized = []

# Loop through each document in your 'docs' list
for doc in docs:
    # Remove punctuations and keep only alphabetic characters
    doc = re.sub(r'[^a-zA-Z]','', doc)
    # Tokenize the document into words
    words = doc.split()
    # Lemmatize and remove stopwords from the document
    filtered_words = [lemmatizer.lemmatize(word.lower()) for word in words if word.lower() not in stop_words]
    # Join the filtered words back into a document
    filtered_doc = ' '.join(filtered_words)
    # Append the document without stopwords and with lemmatized words to the new list
    docs_without_stopwords_lemmatized.append(filtered_doc)

# Now 'docs_without_stopwords_lemmatized' contains the documents with stopwords removed and words lemmatized
```

```python
topics,probs=model.fit_transform(docs_without_stopwords)
```
